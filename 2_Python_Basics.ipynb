## 8) List Data Type:

If we want to represent a group of values as a single entity where insertion order
required to preserve and duplicates are allowed then we should go for list data
type.

1) Insertion Order is preserved

2) Heterogeneous Objects are allowed

3) Duplicates are allowed

4) Growable in nature

5) Values should be enclosed within square brackets.

```python
list=[10,10.5,'Data Science',True,10]
print(list) # [10,10.5,'Data Science',True,10] 
```

```python
# New item assignment is possible
list[0]=100
print(list)
```

```python
# It is growable in nature
list.append(599)
print(list)
```

```python
#removal of item is also possible
list.remove(599)
print(list) 
```

```python
'''Note: An ordered, mutable, heterogenous collection of eleemnts is nothing but list,
where duplicates also allowed.'''
list = list*2
print(list)
```

## 11) set Data Type:

֍ If we want to represent a group of values without duplicates where order is not
important then we should go for set Data Type.

1) Insertion order is not preserved

2) Duplicates are not allowed

3) Heterogeneous objects are allowed

4) Index concept is not applicable

5) It is mutable collection

6) Growable in nature

```python
s={100,0,10,200,10,'Data Science'}
```

```python
s[0] # set does not support indexing
```

```python
# set is growable in nature, based on our requirement we can increase or decrease the size.
s.add(60)
s 
```

```python
# We can remove the element from set
s.remove(100) 
s
```

dict data type.
